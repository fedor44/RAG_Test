```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
```

```python
template = '''Answer the question based only on the following context:
{context}
Question: {question}
'''

prompt = ChatPromptTemplate.from_template(template)
model = ChatOpenAI(model="gpt-4o-mini", temperature=0)

retriever = vectorstore.as_retriever()

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

rag_chain = (
    {"context" : retriever | format_docs, "question" : RunnablePassthrough()}
    | prompt
    | model
    | StrOutputParser()
)

rag_chain.invoke("격하 과정에 대해서 알려줘")
```

```python
import numpy as np
from numpy import dot
from numpy.linalg import norm

def cos_sim(A, B):
    return dot(A, B)/(norm(A)*norm(B))
```

```python
sentences = [
    '안녕하세요',
    '어! 오랜만이오',
    '이름은 어케 되세요?',
    '날씨가 추워요',
    'hello llm'
]
```

```python
from langchain_openai import OpenAIEmbeddings

embeddings_model = OpenAIEmbeddings()
open_embedding = embeddings_model.embed_documents(sentences)

print(open_embedding[0])
```

```python
embedded_query = embeddings_model.embed_query("첫인사를 하고 이름을 물어 봤나요?")
embedded_query
```

```python
for embedding in open_embedding:
    print(cos_sim(embedded_query, embedding))
```

```python
from langchain_community.vectorstores import Chroma

loader = TextLoader("/content/rag_docs/history.txt")
data = loader.load()

text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    chunk_size = 250,
    chunk_overlap = 50,
    encoding_name="cl100k_base"
)

texts = text_splitter.split_text(data[0].page_content)
```

```python
embeddings_model = OpenAIEmbeddings()

db = Chroma.from_texts(
    texts,
    embeddings_model,
    collection_name="history",
    persist_directory="./db/chromadb",
    collection_metadata={"hnsw:space":"cosine"}
)
```

```python
query = "누가 한글을 창제 했나요?"
docs = db.similarity_search(query)
print(docs[0].page_content)
```

```python
from langchain_community.vectorstores import FAISS
from langchain_community.vectorstores.utils import DistanceStrategy
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.docstore.document import Document
```

```python
!pip install faiss-cpu
```

```python
embeddings_model = HuggingFaceEmbeddings(
    model_name="jhgan/ko-sroberta-nli",
    encode_kwargs={"normalize_embeddings" : True}
)

documents = []
for i in ["/content/rag_docs/history.txt", "/content/rag_docs/places.txt"]:
  with open(i, "r", encoding='utf-8') as file:
    content = file.read()
    documents.append(Document(page_content=content))

vectorstore = FAISS.from_documents(
    documents,
    embedding=embeddings_model,
    distance_strategy=DistanceStrategy.COSINE
)

vectorstore
```

```python
query = "서울을 대표하는 거리는?"
docs = vectorstore.similarity_search(query)
print(docs[0].page_content)
```
